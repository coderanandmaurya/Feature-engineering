```python
import numpy as np
import pandas as pd
```

```python
raw_csv="https://raw.githubusercontent.com/coderanandmaurya/Feature-engineering/main/train.csv"
df1 = pd.read_csv(raw_csv)
```

```python
df1.head()
```

```python
df=df1[["Cabin"]]
```

```python
df.sample()
```

```python
df['Cabin'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # captures numerical part
df['cabin_cat'] = df['Cabin'].str[0] # captures the first letter

df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```
